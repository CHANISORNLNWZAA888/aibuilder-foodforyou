```python
import pandas as pd
import json
from google.colab import drive
from IPython.display import display
from difflib import get_close_matches

recipe_file_path = "เมนูไม่มีจุดแล้ว.csv"
recipe_data = pd.read_csv(recipe_file_path)

nutrition_file_path = "imgrediemtdnewnew.csv"
nutrition_df = pd.read_csv(nutrition_file_path)

def transform_ingredients(json_str):
    try:
        data = json.loads(json_str)
        recipe_data = list(data.values())[0]
        ingredients = recipe_data.get('วัตถุดิบ', [])
        return {ing['ชื่อ']: ing['ปริมาณ'] for ing in ingredients}
    except:
        return {}

recipe_data['simplified_ingredients'] = recipe_data['ingredients_full_json'].apply(transform_ingredients)
cleaned_recipe_data = recipe_data[['ชื่ออาหาร', 'simplified_ingredients']]

thai_names = nutrition_df['Thai_Name'].astype(str).tolist()

def parse_ingredients(ingredients_str):
    ingredients_list = ingredients_str.split('\n')
    ingredients_dict = {}
    for ingredient_str in ingredients_list:
        try:
            name, quantity = ingredient_str.split(', ', 1)
            ingredients_dict[name.strip()] = quantity.strip()
        except ValueError:
            print(f"Skipping: {ingredient_str}")
    return ingredients_dict

def find_closest_match(ingredient, thai_names):
    matches = get_close_matches(ingredient, thai_names, n=1, cutoff=0.3)
    return matches[0] if matches else None

def calculate_calories(fat, protein, carbs):
    return (fat * 9) + (protein * 4) + (carbs * 4)

for index, row in cleaned_recipe_data.iterrows():
    food_name = row['ชื่ออาหาร']
    ingredients = row['simplified_ingredients']
    total_calories = 0
    for ingredient, quantity in ingredients.items():
        closest_match = find_closest_match(ingredient, thai_names)
        if closest_match:
            nutrition_info = nutrition_df[nutrition_df['Thai_Name'] == closest_match]
            if not nutrition_info.empty:
                fat = float(nutrition_info['Fat(g)'].iloc[0])
                protein = float(nutrition_info['Protein(g)'].iloc[0])
                if nutrition_info['Carbs(g)'].iloc[0] == '-' or nutrition_info['Carbs(g)'].iloc[0] == '':
                  carbs = 0
                else:
                  carbs = float(nutrition_info['Carbs(g)'].iloc[0])

                calories = calculate_calories(fat, protein, carbs)
                total_calories += calories
                print(f"Ingredient: {ingredient}, Closest Match: {closest_match}, Calories: {calories}")

            else:
                print(f"No nutritional information found for {closest_match}")
        else:
            print(f"No close match found for ingredient: {ingredient}")
    print(f"Total calories for {food_name}: {total_calories}")
```
